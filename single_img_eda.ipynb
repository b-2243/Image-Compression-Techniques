```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('jpeg_43-2.jpg')

```

```python
img
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()
```

```python
mean = np.mean(img)
median = np.median(img)
std = np.std(img)

print(f'Mean: {mean:.2f}, Median: {median:.2f}, Std: {std:.2f}')
```

```python
b, g, r = cv2.split(img)
```

```python
b
```

```python
# plt.subplot(nrows, ncols, index)
# nrows: Number of rows in the subplot grid
# ncols: Number of columns in the subplot grid
# index: Index of the subplot (starts from 1)
```

```python
b, g, r = cv2.split(img)
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1), plt.imshow(b, 'gray'), plt.title('Blue Channel')
plt.subplot(1, 3, 2), plt.imshow(g, 'gray'), plt.title('Green Channel')
plt.subplot(1, 3, 3), plt.imshow(r, 'gray'), plt.title('Red Channel')
plt.show()
```

```python
img.flatten()
```

```python
plt.hist(img.flatten(), bins=256)
plt.title('Histogram of Pixel Values')
plt.show()
```

```python
_, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_BGR2RGB))
plt.title('Thresholded Image')
plt.show()
```

```python
# Lower threshold (100): Pixels with values below this threshold are considered non-edges.
# Upper threshold (200): Pixels with values above this threshold are considered edges.
```

```python
# Step 6: Visualize Edges
edges = cv2.Canny(img, 100, 200)
plt.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB))
plt.title('Edges')
plt.show()
```
